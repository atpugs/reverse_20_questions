```python
from sentence_transformers import SentenceTransformer
sent_model = SentenceTransformer('paraphrase-distilroberta-base-v1')

# For the DPR
passage_encoder = SentenceTransformer('facebook-dpr-ctx_encoder-single-nq-base')
query_encoder = SentenceTransformer('facebook-dpr-question_encoder-single-nq-base')

# For Google's NQ DistilBERT
nq_model = SentenceTransformer('nq-distilbert-base-v1')
```

```python
import requests
import nltk
from ast import literal_eval
from gensim import corpora
from gensim.summarization import bm25
from pprint import pprint
# from scipy.spatial.distance import cdist
from sentence_transformers import util

WIKIMEDIA_API = "https://en.wikipedia.org/w/api.php?format=json&action=query&prop=extracts&explaintext&redirects=1&titles={}&origin=*"

def fetch_wiki_article(name):
    api = WIKIMEDIA_API.format(name)
    content = requests.get(api)
    pages = literal_eval(content.text)["query"]["pages"]
    extracts = list()
    for page_key in pages.keys():
        extract = pages[page_key]["extract"]
        title = pages[page_key]["title"]
        extracts += (extract, title)
    return extracts

def tokenize_sentences(text):
    sent_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
    return sent_tokenizer.tokenize(text.strip())

def rank_bm25(query, sentences, topN):
    sentence_splits = [sent.split() for sent in sentences]
    dictionary = corpora.Dictionary(sentence_splits)
    corpus = [dictionary.doc2bow(token) for token in sentence_splits]
    bm25_obj = bm25.BM25(corpus)
    query_doc = dictionary.doc2bow(query.split())
    scores = bm25_obj.get_scores(query_doc)
    best_docs = sorted(range(len(scores)), key=lambda i: scores[i])[-topN:]
    return [sentences[i] for i in best_docs]

def fetch_top_n_nq(title, query, N=100):
    article, title = fetch_wiki_article(title)
    sentences = tokenize_sentences(article)
    # print(rank_bm25(query, sentences, 5))
    query_embedding = nq_model.encode(query)
    ranks = list()
    for i, sent in enumerate(rank_bm25(query, sentences, N)):
        sentence_embedding = nq_model.encode([[title, sent]])
        score = util.pytorch_cos_sim(query_embedding, sentence_embedding)
        ranks.append((score, sent))
    return ranks

def fetch_top_n_dpr(title, query, N=100):
    article, title = fetch_wiki_article(title)
    sentences = tokenize_sentences(article)
    # print(rank_bm25(query, sentences, 5))
    query_embedding = query_encoder.encode(query)
    ranks = list()
    for i, sent in enumerate(rank_bm25(query, sentences, N)):
        sentence_embedding = passage_encoder.encode(title + " [SEP] " + sent)
        score = util.dot_score(query_embedding, sentence_embedding)
        ranks.append((score, sent))
    return ranks

def fetch_top_n(title, query, N=100):
    article, title = fetch_wiki_article(title)
    sentences = tokenize_sentences(article)
    # print(rank_bm25(query, sentences, 5))
    query_embedding = sent_model.encode(query)
    ranks = list()
    for i, sent in enumerate(rank_bm25(query, sentences, N)):
        sentence_embedding = sent_model.encode(sent)
        score = util.dot_score(query_embedding, sentence_embedding)
        ranks.append((score, sent))
    return ranks

scores = fetch_top_n_nq("Cheetah", "Can it be used in shows?")
```

```python
from copy import deepcopy
import akinator
```

```python
def play_baseline_game(guess):
    aki = akinator.Akinator()
    
    choices = ["yes", "no", "idk", "probably", "probably not"]
    choice_prob = [0.425, 0.425, 0.05, 0.05, 0.05]
    
    q = aki.start_game(language="en_animals")
    prob_guess = 0
    feature_count = 0
    
    while aki.progression <= 80:
        aug_guess = "it"
        question = aki.question.replace("your animal", aug_guess)
        scores = fetch_top_n_nq(guess, question)
        print([(score[0].item(), sent) for score, sent in sorted(scores, key=lambda x: x[0].item(), reverse=True)[:2]])
        human_reply = input(question + "\n\t")
        print("Question: ", aki.question)
        print("Answer: ", human_reply)
        q = aki.answer(human_reply)
        prev_aki = deepcopy(aki)
        prev_aki.win()
        print([(_guess['name'], _guess['proba']) for _guess in prev_aki.guesses])
        for g in prev_aki.guesses:
            if g['name'] in guess or guess in g['name']:
                prob_guess += float(g['proba'])
                feature_count += 1
                break
    aki.win()
    
    if feature_count == 0:
        print("Avg probability: ", 0)
    else:
        print("Avg probability: ", prob_guess/feature_count)
    print("I guess", aki.first_guess['name'])
    if guess in aki.first_guess['name']:
        print("Yay\n")
    else:
        print("Oof\n")
```
